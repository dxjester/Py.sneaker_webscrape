# Plot Snippets

```python
# line chart
import matplotlib.pyplot as plt
x_vals = [x for x in range(10)]
y_vals = [x**3 for x in range (10)]

plt.plot(x_vals,y_vals)
plt.show()
```

```python
# scatter chart
import matplotlib.pyplot as plt
x_vals = [x for x in range(25)]
y_vals = [x**0.5 for x in range (25)]

plt.scatter(x_vals,y_vals)
plt.show()
```

```python
# line chart
import matplotlib.pyplot as plt

y_vals = [x*x for x in range (10)]

plt.boxplot(y_vals)
plt.show()
```

```python
# tree map
import matplotlib.pyplot as plt
import squarify    # pip install squarify (algorithm for treemap)
 
# If you have 2 lists
squarify.plot(sizes=[13,22,35,5], label=["group A", "group B", "group C", "group D"], alpha=.7 )
plt.axis('off')
plt.show()
 
# If you have a data frame?
import pandas as pd
df = pd.DataFrame({'nb_people':[8,3,4,2], 'group':["group A", "group B", "group C", "group D"] })
squarify.plot(sizes=df['nb_people'], label=df['group'], alpha=.8 )
plt.axis('off')
plt.show()
```
